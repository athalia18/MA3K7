```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def bracelet(a,b,N):
    chain = [a,b]
    for i in range(2,N):
        chain.append((chain[i-2]+chain[i-1])%10)
    return chain

def arra(N):
    arr = []
    for i in range(N):
        arr.append(i+1)
    return arr  
```

```python
x1 = arra(40)
y1 = bracelet(1,5,40)
y2 = bracelet(7,8,40)
plt.plot(x1,y1,x1,y2)
plt.legend([r'(1,5)',r'(7,8)'])
plt.title('40 Terms in Chain')
plt.xlabel('$\mathregular{n^{th} term}$')
plt.ylabel('$\mathregular{a_n value}$')
plt.show()
```
