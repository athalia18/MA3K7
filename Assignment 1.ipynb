```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def bracelet(a,b,N):
    chain = [a,b]
    for i in range(2,N):
        chain.append((chain[i-2]+chain[i-1])%10)
    return chain

def arra(N):
    arr = []
    for i in range(N):
        arr.append(i+1)
    return arr  
```

```python
x1 = arra(40)
y1 = bracelet(1,5,40)
y2 = bracelet(7,8,40)
plt.plot(x1,y1,x1,y2)
plt.legend([r'(1,5)',r'(7,8)'])
plt.title('40 Terms in Chain')
plt.xlabel('$\mathregular{n^{th} term}$')
plt.ylabel('$\mathregular{a_n value}$')
plt.grid()
plt.show()
```

```python
x1 = arra(2000)
y1 = bracelet(1,5,2000)
y2 = bracelet(7,8,2000)
plt.plot(x1,y1,x1,y2)
plt.legend([r'(1,5)',r'(7,8)'])
plt.title('2000 Terms in Chain')
plt.xlabel('$\mathregular{n^{th} term}$')
plt.ylabel('$\mathregular{a_n value}$')
plt.show()
```

```python
def bracelet1(a,b,N):
    a = a%2
    b = b%2
    chain = [a,b]
    for i in range(2,N):
        chain.append((chain[i-2]+chain[i-1])%2)
    return chain

x1 = arra(15)
y1 = bracelet1(0,0,15)
y2 = bracelet1(1,0,15)
y3 = bracelet1(1,1,15)
y4 = bracelet1(0,1,15)

plt.plot(x1,y1,x1,y2,x1,y3,x1,y4)
plt.legend([r'(0,0)',r'(1,0)',r'(1,1)',r'(0,1)'])
plt.title('15 Terms in Chain using modulo 2')
plt.xlabel('$\mathregular{n^{th} term}$')
plt.ylabel('$\mathregular{a_n value}$')
plt.grid()
plt.show()
```
